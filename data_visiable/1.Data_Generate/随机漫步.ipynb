# 随机漫步类

```python
import matplotlib.pyplot as plt
from randomwalk import RandomWalk

while True:
    rw = RandomWalk()
    rw.fill_walk()

    plt.scatter(
        rw.x_values, rw.y_values,
        c=rw.x_values,
        cmap=plt.cm.Blues,
        edgecolors=None,
        s=15
    )

    plt.show()

    running_flag = input("Wanna another round?(y/n): ")
    if running_flag.lower() == 'n':
        break
```

## 1. 绘制起点和终点

```python
import matplotlib.pyplot as plt
from randomwalk import RandomWalk

rw = RandomWalk()
rw.fill_walk()


# 突出显示起点
plt.scatter(
    0, 0,
    c='green',
    edgecolors=None,
    s=400)

# 突出显示终点
plt.scatter(
    rw.x_values[-1], rw.y_values[-1],
    c='red',
    edgecolors=None,
    s=400)

plt.scatter(
    rw.x_values, rw.y_values,
    c=rw.x_values,
    cmap=plt.cm.Blues,
    edgecolors=None,
    s=15
)

plt.show()
```

## 2. 隐藏标尺

```python
import matplotlib.pyplot as plt
from randomwalk import RandomWalk
%matplotlib inline

rw = RandomWalk()
rw.fill_walk()


# 突出显示起点
plt.scatter(
    0, 0,
    c='green',
    edgecolors=None,
    s=400)

# 突出显示终点
plt.scatter(
    rw.x_values[-1], rw.y_values[-1],
    c='red',
    edgecolors=None,
    s=400)

plt.scatter(
    rw.x_values, rw.y_values,
    c=rw.x_values,
    cmap=plt.cm.Blues,
    edgecolors=None,
    s=15
)

# 隐藏坐标轴
plt.axis('off')

plt.show()
```

## 3.改变图像大小

```python
import matplotlib.pyplot as plt
from randomwalk import RandomWalk
%matplotlib inline

rw = RandomWalk(50000)
rw.fill_walk()

# 对应的每一个坐标点的顺序，用于后面颜色映射
point_num = list(range(rw.num_points))

# 改变显示图框的大小
plt.figure(dpi=128,figsize=(10, 6))

# 突出显示起点
plt.scatter(
    0, 0,
    c='green',
    edgecolors=None,
    s=100)

# 突出显示终点
plt.scatter(
    rw.x_values[-1], rw.y_values[-1],
    c='red',
    edgecolors=None,
    s=100)

plt.scatter(
    rw.x_values, rw.y_values,
    c=point_num,
    cmap=plt.cm.Blues,
    edgecolors=None,
    s=1
)


# 隐藏坐标轴
plt.axis('off')

plt.show()
```
