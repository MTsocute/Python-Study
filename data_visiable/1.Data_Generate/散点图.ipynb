# 绘制散点图

```python
import matplotlib.pyplot as plt

# s = size
# plt.scatter(2, 4, s=200)   # 绘制点个点（x,y）

def scatter_demo():
    x_values = [1, 2, 3, 4, 5]
    y_values = [1, 4, 9, 16, 25]

    plt.scatter(x_values, y_values,s=100)

    plt.title("Square Numbers", fontsize=24)
    plt.xlabel("Value", fontsize=14)
    plt.ylabel("Square of Value", fontsize=14)

    # major: 标尺不会很细，刻度间隙会粗略些
    plt.tick_params(axis='both', which='major', labelsize=14)
    plt.show()

scatter_demo()
```

```python
def scatter_demo_2():
    x_val = list(range(1, 1001))
    y_val = [pow(x,2) for x in x_val]   # 所有的点都是 x 的平方倍

    plt.figure(figsize=(8, 6))  # 设置图表大小

    plt.scatter(x_val, y_val, s=10)

    # 设置坐标轴的取值范围
    plt.axis(
        [0,1100,    # x 轴
        0,1.1e6]    # y 轴
    )
    plt.show()

scatter_demo_2()
```

## 使用特定的颜色来绘制散点图像

~~~python
    # 直接说颜色
    plt.scatter(x_values, y_values, c='red', edgecolor='none', s=40)
    # 利用 rgb 来调色
    """ 对应 Red Green Blue 但是在 0~1 之间取值，越接近0越深，1越浅 """
    plt.scatter(x_values, y_values, c=(0, 0, 0.8), edgecolor='none', s=40)
~~~

```python
import matplotlib.pyplot as plt
def scatter_demo_3():
    x_val = list(range(1, 1001))
    y_val = [pow(x,2) for x in x_val]   # 所有的点都是 x 的平方倍

    plt.figure(figsize=(8, 6))  # 设置图表大小

    # 指定点的颜色为红色
    plt.scatter(x_val, y_val,
                c='red',    # 绘制点的颜色
                edgecolor='none',   # 删除点的外边框颜色
                s=40)

    # 设置坐标轴的取值范围
    plt.axis(
        [0,1100,    # x 轴
        0,1.1e6]    # y 轴
    )
    plt.show()

scatter_demo_3()
```

## 使用颜色渐变

```python
import matplotlib.pyplot as plt
def scatter_demo_4():
    x_val = list(range(1, 1001))
    y_val = [pow(x,2) for x in x_val]   # 所有的点都是 x 的平方倍

    plt.figure(figsize=(8, 6))  # 设置图表大小

    # 颜色渐变
    plt.scatter(x_val, y_val,
                c=y_val,    # 根据每个点的 y 值来改变颜色
                cmap=plt.cm.Blues,  # 颜色渐变
                edgecolor='none', s=40)

    # 设置坐标轴的取值范围
    plt.axis(
        [0,1100,    # x 轴
        0,1.1e6]    # y 轴
    )
    plt.show()

scatter_demo_4()
```

```python
import matplotlib.pyplot as plt
def scatter_demo_5():
    """ 彩色渐变 """
    x_val = list(range(1, 1001))
    y_val = [pow(x,2) for x in x_val]   # 所有的点都是 x 的平方倍

    plt.figure(figsize=(8, 6))  # 设置图表大小

    # 颜色渐变
    plt.scatter(x_val, y_val,
                c=y_val,    # 根据每个点的 y 值来改变颜色
                cmap='rainbow',  # 颜色渐变
                edgecolor='none', s=40)

    # 设置坐标轴的取值范围
    plt.axis(
        [0,1100,    # x 轴
        0,1.1e6]    # y 轴
    )
    plt.show()

scatter_demo_5()
```
