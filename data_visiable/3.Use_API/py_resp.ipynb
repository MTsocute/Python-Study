## 请求和请求对象生成

```python
import requests

url = 'https://api.github.com/search/repositories?q=language:python&sort=stars' # 请求地址
r = requests.get(url)   # 请求对象
print(r)
print(f'Status Code: {r.status_code}')  # 响应状态码
```

## 细节研究下内容

```python
import requests

url = 'https://api.github.com/search/repositories?q=language:python&sort=stars' # 请求地址
r = requests.get(url)   # 请求对象

# 将API响应存储在一个字典中
rep_dict = r.json()

# 多少个仓库
print("Total repositories:", rep_dict['total_count'])   

# 有关仓库的信息
repo_dicts = rep_dict['items']
print("Repositories returned:", len(repo_dicts))

# 研究第一个仓库
repo_dict = repo_dicts[0]
print("\nKeys:", len(repo_dict))
for key in sorted(repo_dict.keys()):
    print(key)
```

```python
import requests

url = 'https://api.github.com/search/repositories?q=language:python&sort=stars' # 请求地址
r = requests.get(url)   # 请求对象

# 将API响应存储在一个字典中
rep_dict = r.json()

# 多少个仓库
print("Total repositories:", rep_dict['total_count'])   

# 有关仓库的信息
repo_dicts = rep_dict['items']
print("Repositories returned:", len(repo_dicts))

# 研究第一个仓库
print("\nSelected information about first repository:")
repo_dict = repo_dicts[0]

print('ID:', repo_dict['id'])
print('Name:', repo_dict['name'])
print('Owner:', repo_dict['owner']['login'])
print('Stars:', repo_dict['stargazers_count'])
print('Repository:', repo_dict['html_url'])
print('Created:', repo_dict['created_at']) 
print('Updated:', repo_dict['updated_at'])
print('Description:', repo_dict['description'])
```

## 遍历每个仓库描述基本情况

```python
import requests

url = 'https://api.github.com/search/repositories?q=language:python&sort=stars' # 请求地址
r = requests.get(url)   # 请求对象

# 将API响应存储在一个字典中
rep_dict = r.json()

# 多少个仓库
print("Total repositories:", rep_dict['total_count'])

# 有关仓库的信息
repo_dicts = rep_dict['items']
print("Repositories returned:", len(repo_dicts))

# 遍历每一个仓库
print("\nSelected information about each repository:")
for rep in repo_dicts[:100]:
    print('ID:', rep['id'])
    print('Name:', rep['name'])
    print('Owner:', rep['owner']['login'])
    print('Stars:', rep['stargazers_count'])
    print('Repository:', rep['html_url'])
    print('Created:', rep['created_at'])
    print('Updated:', rep['updated_at'])
    print('Description:', rep['description'])
    print("-"*66)
```
