```python
import requests

url = 'https://api.github.com/search/repositories?q=language:python&sort=stars' # ËØ∑Ê±ÇÂú∞ÂùÄ
r = requests.get(url)   # ËØ∑Ê±ÇÂØπË±°

# Â∞ÜAPIÂìçÂ∫îÂ≠òÂÇ®Âú®‰∏Ä‰∏™Â≠óÂÖ∏‰∏≠
rep_dict = r.json()

# Â§öÂ∞ë‰∏™‰ªìÂ∫ì
print("Total repositories:", rep_dict['total_count'])

# ÊúâÂÖ≥‰ªìÂ∫ìÁöÑ‰ø°ÊÅØ
repo_dicts = rep_dict['items']
print("Repositories returned:", len(repo_dicts))

# ÈÅçÂéÜÊØè‰∏Ä‰∏™‰ªìÂ∫ì
print("\nSelected information about each repository:")
for rep in repo_dicts[:100]:
    print('ID:', rep['id'])
    print('Name:', rep['name'])
    print('Owner:', rep['owner']['login'])
    print('Stars:', rep['stargazers_count'])
    print('Repository:', rep['html_url'])
    print('Created:', rep['created_at'])
    print('Updated:', rep['updated_at'])
    print('Description:', rep['description'])
    print("-"*66)
```
