```python
import requests
from operator import itemgetter

url = 'https://hacker-news.firebaseio.com/v0/topstories.json'   # 热门文章的ID
r = requests.get(url)
print("Status code:", r.status_code)

# 处理每篇文章的信息
submission_ids = r.json()   # 返回一个列表全部都是热门文章的 ID
submission_dicts = []       # 存储所需资料的大列表
for submission_id in submission_ids[:30]:
    # 对每一篇文章都进行对应相关消息的 API 调用
    url = ('https://hacker-news.firebaseio.com/v0/item/'
           + str(submission_id) + '.json')  # 请求API
    submission_r = requests.get(url)
    print(submission_r.status_code)         # 每个请求的状态码
    response_dict = submission_r.json()     # 每个返回数据都是字典

    # 存储我们需要的数据
    submission_dict = {
        'title': response_dict['title'],
        'link': 'http://news.ycombinator.com/item?id=' + str(submission_id),
        # .get() 返回对应的值，0 是默认值，如果不存在就是返回 0
        'comments': response_dict.get('descendants', 0)
    }

    # 存储我们整理的数据在大列表中
    submission_dicts.append(submission_dict)

# 看看内容
print(submission_dicts)
```

## 利用模块自动排序

~~~python
    submission_dicts = sorted(submission_dicts, key=itemgetter('comments'), reverse=True)
~~~
> 只要往 itemgetter()里，传入我们需要的键，就会捕捉对应的值，然后我们可以利用这个来实现排序

```python
import requests
from operator import itemgetter

url = 'https://hacker-news.firebaseio.com/v0/topstories.json'   # 热门文章的ID
r = requests.get(url)   # 请求热门文章ID的请求体
print("Status code:", r.status_code)

# 处理每篇文章的信息
submission_ids = r.json()   # 返回一个列表全部都是热门文章的 ID 的 json 数据
submission_dicts = []       # 存储所需资料的大列表
for submission_id in submission_ids[:30]:
    # 对每一篇文章都进行对应相关消息的 API 调用
    url = ('https://hacker-news.firebaseio.com/v0/item/'
           + str(submission_id) + '.json')  # 请求API
    submission_r = requests.get(url)
    print('Sub Status: ' + str(submission_r.status_code))    # 每个请求的状态码
    response_dict = submission_r.json()     # 每个返回数据都是字典

    # 存储我们需要的数据
    submission_dict = {
        'title': response_dict['title'],
        'link': 'http://news.ycombinator.com/item?id=' + str(submission_id),
        # .get() 返回对应的值，0 是默认值，如果不存在就是返回 0
        'comments': response_dict.get('descendants', 0)
    }

    # 存储我们整理的数据在大列表中
    submission_dicts.append(submission_dict)

# 按照评论数降序
submission_dicts = sorted(submission_dicts,
                          key=itemgetter('comments'), reverse=True)

# 展示内容
for submission_dict in submission_dicts:
    print("\nTitle:", submission_dict['title'])
    print("Discussion link:", submission_dict['link'])
    print("Comments:", submission_dict['comments'])
```

## 可视化

```python
import pygal
import requests
from operator import itemgetter

url = 'https://hacker-news.firebaseio.com/v0/topstories.json'   # 热门文章的ID
r = requests.get(url)   # 请求热门文章ID的请求体
print("Status code:", r.status_code)

# 处理每篇文章的信息
submission_ids = r.json()   # 返回一个列表全部都是热门文章的 ID 的 json 数据
submission_dicts = []       # 存储所需资料的大列表
for submission_id in submission_ids[:30]:
    # 对每一篇文章都进行对应相关消息的 API 调用
    url = ('https://hacker-news.firebaseio.com/v0/item/'
           + str(submission_id) + '.json')  # 请求API
    submission_r = requests.get(url)
    print('Sub Status: ' + str(submission_r.status_code))    # 每个请求的状态码
    response_dict = submission_r.json()     # 每个返回数据都是字典

    # 存储我们需要的数据
    submission_dict = {
        'title': response_dict['title'],
        'link': 'http://news.ycombinator.com/item?id=' + str(submission_id),
        # .get() 返回对应的值，0 是默认值，如果不存在就是返回 0
        'comments': response_dict.get('descendants', 0)
    }

    # 存储我们整理的数据在大列表中
    submission_dicts.append(submission_dict)

# 按照评论数降序
submission_dicts = sorted(submission_dicts,
                          key=itemgetter('comments'), reverse=True)



""" 可视化 """
titles, plot_dicts= [], []
for submission_dict in submission_dicts:
    titles.append(submission_dict['title'])

    plot_dict = {
        'value': submission_dict.get('comments', 0),
        'xlink': submission_dict['link']
    }

    plot_dicts.append(plot_dict)

# 优化下设置，让图更好看
my_config = pygal.Config()
my_config.x_label_rotation = 45     # 坐标标签旋转 45 度
my_config.show_legend = False       # 不显示图例(左上角小按钮)
my_config.title_font_size = 24
my_config.label_font_size = 14
my_config.major_label_font_size = 18
my_config.truncate_label = 15       # 如果坐标标签长度超过15个字符，则截断显示
my_config.show_y_guides = False     # 不显示Y轴的辅助线
my_config.width = 1000

chart = pygal.Bar(my_config)

# 横坐标标签
chart.x_labels = titles

chart.add(' ', plot_dicts)

chart.render_to_file('/Users/momo/Desktop/Python/data_visiable/images/hacker.svg')

print("Done!")
```
